```python
import pandas as pd
import os

from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly import graph_objs as go
import requests
import pandas as pd

print(__version__) # need 1.9.0 or greater

init_notebook_mode(connected = True)

def plotly_df(df, title = ''):
    data = []
    
    for column in df.columns:
        trace = go.Scatter(
            x = df.index,
            y = df[column],
            mode = 'lines',
            name = column
        )
        data.append(trace)
    
    layout = dict(title = title)
    fig = dict(data = data, layout = layout)
    iplot(fig, show_link=False)
```

## Предсказание с помощью Facebook Prophet

Для начала построим предсказание с помощью простой библиотеки `Facebook Prophet`. Для того, чтобы посмотреть на качество модели, отбросим из обучающей выборки последние 30 дней.

```python
from fbprophet import Prophet
```

```python
predictions = 30

df = df[['date', 'count']]
df.columns = ['ds', 'y']
train_df = df[:-predictions].copy()
```

```python
train_df = df[:-predictions].copy()
```

```python
prophet_model = Prophet(yearly_seasonality=True, daily_seasonality=True)
prophet_model.fit(train_df)
```

```python
future = prophet_model.make_future_dataframe(periods=predictions)
future.tail()
```

```python
forecast = prophet_model.predict(future)
forecast.tail()
```

```python
prophet_model.plot_components(forecast)
```

```python
df = forecast.set_index('ds')[['yhat', 'yhat_lower', 
                                   'yhat_upper']].join(df.set_index('ds'))
```

```python
from numpy import mean
df['e'] = df['y'] - df['yhat']
df['p'] = 100*df['e']/df['y']
print('MAPE = ', round(mean(abs(df[-predictions:]['p'])), 2))
print('MAE = ', round(mean(abs(df[-predictions:]['e'])), 2))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

**Вопрос 4:** Проверим стационарность ряда с помощью критерия Дики-Фулера. Является ли ряд стационарным? Какое значение p-value?

Стационарный будет только на сезонах, по годам p=0.107392, по сезону p=0

```python
sm.tsa.seasonal_decompose(train_df['y'].values, period=7).plot();
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(train_df['y'])[1])
```

```python
train_df.set_index('ds', inplace=True)
```

```python
train_df['y_diff'] = train_df.y - train_df.y.shift(7)
sm.tsa.seasonal_decompose(train_df.y_diff[7:].values, period=7).plot();
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(train_df.y_diff[8:])[1])
```

```python
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(train_df.y_diff[13:].values.squeeze(), lags=48, ax=ax)

ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(train_df.y_diff[13:].values.squeeze(), lags=48, ax=ax)
```
